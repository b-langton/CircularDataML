```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import scipy
```

```python
##Define equations of motion for the pivot point: 
##We will change these as time goes on to explore
def Y(t): 
    return  .1*250*250*np.cos(250*t)
def X(t): 
    return 0

```

```python
##define the function that calculates the derivatives of the things we are trying to solve
def f(y, t):
    L = 1
    theta, omega = y
    return [omega, -1*(9.81 - Y(t))*np.sin(theta)/L - X(t)*np.cos(theta)/L]

##define the time interval and points to check
end = 20
interval = .01
t = np.arange(0, end, interval)
y0 = [np.pi, 0.01] ##initial conditions for the ODE
other = [1] ##defining L
sln = odeint(f, y0, t, hmax = .005)
    
```

```python
##plot theta vs time
fig = plt.figure(figsize = (10, 10))
ax1 = fig.add_subplot(311)
ax1.plot(t, sln[:,0])
ax1.set_xlabel('time')
ax1.set_ylabel('theta')
ax1.set_title("Theta vs Time for inverted Pendulum, high frequency vertical input")
```
