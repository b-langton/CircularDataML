```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint, solve_ivp
import scipy
import math
```

```python
##Define equations of motion (second derivative) for the pivot point: 
##We will change these as time goes on to explore
def Y(t): 
    return  .1*250*250*np.cos(250*t)
def X(t): 
    return .1*250*100*np.sin(250*t)

```

```python
##define the function that calculates the derivatives of the things we are trying to solve
def f(y, t):
    L = 1
    theta, omega = y
    return [omega, -1*(9.81 - Y(t))*np.sin(theta)/L - X(t)*np.cos(theta)/L]

##define the time interval and points to check
end = 20
interval = .01
t = np.arange(0, end, interval)
y0 = [np.pi, 0.01] ##initial conditions for the ODE
other = [1] ##defining L
sln = odeint(f, y0, t, hmax = .005)
    
```

```python
##plot theta vs time
fig = plt.figure(figsize = (10, 10))
ax1 = fig.add_subplot(311)
ax1.plot(t, sln[:,0])
ax1.set_xlabel('time')
ax1.set_ylabel('theta')
ax1.set_title("Theta vs Time for inverted Pendulum, high frequency horizontal input")
```

```python
##Code for the 3d pendulum. This is what mathematica gives for our two governing equations ( theta and phi respectively)
## L*m*(g*Sin(theta) - .5*L*Sin(2*theta)*phi'*phi' + Cos(theta)Cos(phi)*Fx(t) + Cos(theta)Sin(phi)Fy(t) + Sin(theta)\
## *Fz[t] + Ltheta'') = 0
## LmSin(theta)*(2*L*Cos(theta)*omega*phi' - Sin(phi)*Fx(t) + Cos(phi)*Fy(t) + L*Sin(theta)*phi'')


## define helper functions
def Sin(x): 
    return math.sin(x)
def Cos(x): 
    return math.cos(x)


##These actually represent the 2nd derivatives of the x, y, and z locations of the pivot point.
def Fx(x): 
    return 0
def Fy(x): 
    return 0
def Fz(x): 
    return  .1*250*250*np.cos(250*x)
    
    
    
def f(y,t): 
    L = 1
    m = 1
    g = 9.8
    theta, omega, phi, upsilon = y 
    ##This is the system of DEs that describe the 3d equation of motion that we derived in mathematica. 
    return [omega,-1/L*(g*Sin(theta) - .5*L*Sin(2*theta)*upsilon*upsilon + Cos(theta)*Cos(phi)*Fx(t) + Cos(theta)*Sin(phi)*Fy(t) + Sin(theta)
        *Fz(t)), upsilon, (-1/(L*Sin(theta)))*(2*L*Cos(theta)*omega*upsilon - Sin(phi)*Fx(t) + Cos(phi)*Fy(t))]

end = 20
interval = .01
t = np.arange(0, end, interval)
y0 = [np.pi + .1, 0.5, 0, .5] ##initial conditions for the ODE
other = [1] ##defining L
sln = odeint(f, y0, t, hmax = .0005)
```
